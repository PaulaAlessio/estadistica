```python
"""
Importamos los módulos de python necesarios para nuestros cálculos
"""
import matplotlib.pyplot as plt  # Módulo para poder hacer gráficas
import pandas as pd              # Módulo para tratar datos de manera estadística
import numpy as np               # Módulo para manipular datos
```

```python
"""
Realizamos un diagrama de cajas con bigotes para los datos que contiene el fichero: calificaciones_opo.csv
Los datos corresponden al ejercicio XXX de la página YYY del libro de texto.
Realiza el diagrama de cajas a mano
"""
tabla_de_frecuencias= pd.read_csv("calificaciones_opo.csv")
notas = pd.DataFrame(np.repeat(tabla_de_frecuencias.values[:,0], tabla_de_frecuencias.values[:,1]),columns=["Nota"])
notas.boxplot()
plt.show()
```
